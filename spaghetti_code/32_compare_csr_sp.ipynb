# Compare CSR

```python
rng = np.random.default_rng(0)

num_mics = 16
num_sources = 32
num_active = 3

A = rng.standard_normal((num_mics, num_sources)) + 1j * rng.standard_normal((num_mics, num_sources))

X_true = np.zeros((num_sources, 1), dtype=np.complex128)
active = rng.choice(num_sources, size=num_active, replace=False)
X_true[active, 0] = rng.standard_normal(num_active) + 1j * rng.standard_normal(num_active)

Y = A @ X_true
X0 = np.linalg.pinv(A) @ Y
```

```python
%%time
X_vec = vectorized_sp(Y, A)
```

```python
%%time
X_dense, B_dense = dense_sp(X0, A)
```

```python
%%time
X_scipy = scipy_sp(X0, A, grouping="none", precision=1.0, eps=0.005)
```
