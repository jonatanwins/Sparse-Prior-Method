```python
import numpy as np
from cs_priors.simulation.mixing_model import quick_sim
from cs_priors.solvers.frequency_lasso import frequency_lasso_solve
from cs_priors.solvers.frequency_group_lasso import frequency_group_lasso_solve
from cs_priors.solvers.multifrequency_sparse_prior import sparse_prior_solve
```

```python
sim = quick_sim(num_mics=3, num_sources=4, num_active=2, seed=0, mode="noise", min_freq_hz=1.0)
F = 2
keep = np.flatnonzero(sim.freqs > 0)[:F]

sim.A = sim.A[:, :, keep]
sim.Y = sim.Y[:, keep]
sim.X = sim.X[:, keep]
sim.X_pinv = sim.X_pinv[:, keep]
sim.freqs = sim.freqs[keep]
```

```python
from cs_priors.plotting.plot_complex import plot_matrices


X_lasso = frequency_lasso_solve(sim.Y, sim.A)
X_group_lasso = frequency_group_lasso_solve(sim.Y, sim.A, grouping="frequency")
X_sparse_prior = sparse_prior_solve(sim.X_pinv, sim.A, grouping="frequency") 

plot_matrices([sim.X, sim.X_pinv, X_lasso, X_group_lasso, X_sparse_prior], titles=["True solution X", "Pseudoinverse solution", "Lasso solution", "Group Lasso solution", "Sparse Prior solution"])
```
