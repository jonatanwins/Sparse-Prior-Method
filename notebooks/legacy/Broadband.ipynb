```python
import sys
from pathlib import Path
import os
import numpy as np

# Get the directory where this notebook is located
notebook_dir = Path(os.getcwd())
project_root = notebook_dir.parent  # Go up one level to project root
src_path = project_root / 'src'

if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

from cs_priors.plotting.plotting import (
    plot_equation,
    plot_two_line_equation,
    plot_overview,
    wrapper_plot_geometry,
    plot_matrices,
)
from cs_priors.simulation.mixing_model import run_simulation, just_YAX_from_simulation
from cs_priors.solvers.manual_group_lasso import group_lasso, groups_on_complex_numbers
```

# Broadband signals
From different sources

```python
f0s = [100, 150, 200, 300]
num_sources = len(f0s)
num_mics = int(1.5 * num_sources)
sampling_rate_factor = 3
angle_step = 0.3

sim = run_simulation(
    num_sources=num_sources,
    num_mics=num_mics,
    f0=f0s,
    sampling_rate_factor=sampling_rate_factor,
    angle_step=angle_step,
    )

plot_overview(sim)

plot_equation(sim.Y, sim.A[:, :, 1], sim.X)
```

# Moe Group Lasso

```python
from cs_priors.solvers.moe_group_lasso import frequency_group_lasso, block_vector_to_matrix
P, N, F = sim.A.shape

X_gl_block = frequency_group_lasso(
        sim.A,
        sim.Y,
        num_mics=P,
        num_sources=N,
        num_frequencies=F,
)
X_gl = block_vector_to_matrix(X_gl_block, N, F)

plot_matrices([sim.X, X_gl], titles=["Original X", "Recovered X (Group Lasso)"])
```
